```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y)
plt.title("Sin(x)",loc='center')
plt.grid(visible=True,which='both',axis='x',linestyle='--',linewidth=0.5)
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.cos(x)
plt.plot(x,y)
plt.title("Sin(x)",loc='center')
plt.grid(visible=True,which='both',axis='x',linestyle='--',linewidth=0.5)
plt.show()
```

```python
x = np.linspace(0,10,100)
y = x**2
plt.plot(x,y)
plt.title("Sin(x)",loc='center')
plt.grid(visible=True,which='both',axis='x',linestyle='--',linewidth=0.5)
plt.show()
```

```python
x1 = np.arange(0,10,0.5)
y1 = np.arange(10,20,0.5)
plt.figure(figsize=(3,2), dpi=100)
plt.plot(x1,y1)
plt.xlabel('X Label') # for lable x
plt.ylabel('Y Label') # for lable y
plt.title('Demo Graph') # for title of visualization
plt.show()
```

```python
x1 = np.arange(0,10,0.5)
y1 = np.arange(10,20,0.5)
plt.figure(figsize=(3,2), dpi=100)
plt.plot(x1,y1)
plt.xlabel('X Label') # for lable x
plt.ylabel('Y Label') # for lable y
plt.title('Demo Graph') # for title of visualization
plt.xticks([0,2,4,6,8,10])
plt.yticks([10,12,14,16,18,20])
plt.show()
```

```python
x =[0,1,2,3,4]
y= [0,1,4,6,8]
plt.figure(figsize=(3,2), dpi=100)#set size of chart
# set style of line and marker
plt.plot(x,y,label='y=2X',color='red',linewidth='5', markersize='10',marker='p',linestyle='-', markeredgecolor='b')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('This is title of graph' ,fontdict={'fontname':'Arial' , 'fontsize' : 20 } )
plt.xticks([0,1,2,3])
plt.yticks([0,2,4,6,8,10])

plt.legend() # set lable in graph
plt.show()
```

```python
x =np.arange(0,20,2)
y =np.arange(20,40,2)
plt.figure(figsize=(3,2), dpi=100)#set size of chart
plt.plot(x,y,'gs--',label='2x')  # shorthand notation 'color markerstyle linestyle'
plt.title("This is Line Graph")
plt.legend()
plt.show()
```

```python
x =[0,1,2,3,4]
y= [5,1,4,6,8]
x2=np.arange(0,4,.5)
plt.figure(figsize=(3,2), dpi=100)
# for line one
plt.plot(x,y,'rp--', label='2X')
# for line two
plt.plot(x2, x2**2,'bd--', label='x2')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('This is title of graph' ,fontdict= {'fontname':'Arial' , 'fontsize' : 20 } )
plt.xticks([0,1,2,3])
plt.yticks([0,2,4,6,8,10])
plt.legend()
plt.show()
```

```python
x =[0,1,2,3,4]
y= [0,1,4,6,8]
x2=np.arange(0,4,.5)
plt.figure(figsize=(3,2), dpi=100)
# for line one
plt.plot(x,y,'r.--', label='2X')
# for line two
plt.plot(x2[:5], x2[:5]**2,'b<-', label='x2')
plt.plot(x2[4:], x2[4:]**2,'b>--')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('This is title of graph' ,fontdict= {'fontname':'Arial' , 'fontsize' : 20 } )
plt.xticks([0,1,2,3])
plt.yticks([0,2,4,6,8,10])
plt.legend()
plt.show()
```

```python
x =[0,1,2,3,4]
y= [0,1,4,6,8]
x2=np.arange(0,4,.5)
plt.figure(figsize=(3,2), dpi=100)
plt.plot(x,y,'r.--', label='2X') # for line one
#for line two first part
plt.plot(x2[:5], x2[:5]**2,'b<-', label='x2')
# for line two second part
plt.plot(x2[4:],x2[4:]**2,'b.--')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('This is title of graph' ,fontdict= {'fontname':'Arial' , 'fontsize' : 20 } )
plt.xticks([0,1,2,3])
plt.yticks([0,2,4,6,8,10])
#for save file
plt.savefig("linegraph.jpg",dpi=300)
plt.legend()
plt.show()
```

```python
x =[0,1,2,3,4]
y= [0,1,4,6,8]
x2=np.arange(0,4,.5)
plt.figure(figsize=(3,2), dpi=100)
plt.scatter(x,y, label='2X',color='r',s=20) # for line one
#for line two first part
plt.scatter(x2[:5], x2[:5]**2,color='b', label='x2',alpha=0.5,s=10)
# for line two second part
plt.scatter(x2[4:],x2[4:]**2,color='g')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('This is title of graph' ,fontdict= {'fontname':'Arial' , 'fontsize' : 20 } )
plt.xticks([0,1,2,3])
plt.yticks([0,2,4,6,8,10])

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x1= (np.arange(1,20))
y1= (np.arange(1,20))
plt.figure(figsize=(10,6),dpi=100)
plt.polar(x1,y1,'r.--')
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

theta = np.linspace(0, np.pi, 100)
r = np.sin( theta)                  
plt.polar(theta, r, color='blue', linestyle='-')
plt.title('Polar Plot')
plt.show()
```

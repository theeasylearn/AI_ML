```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-2, 2, 20)
y = np.linspace(-2, 2, 20)
X, Y = np.meshgrid(x, y)
U = -Y
V = X
plt.streamplot(X, Y, U, V, density=1, linewidth=1, color='blue')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Stream Plot')
plt.grid(True)
plt.show()
```

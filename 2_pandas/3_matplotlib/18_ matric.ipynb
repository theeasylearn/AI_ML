```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
A = np.random.rand(5, 5)
plt.matshow(A, cmap='cool')
plt.colorbar(label='Value')
plt.title('Matrix Plot')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x =[0,1,2,3,4]
y= [0,1,4,6,8]
x2=np.arange(0,4,.5)
plt.figure(figsize=(6,2), dpi=100)
plt.scatter(x,y, label='2X',color='r',s=20) # for line one
#for line two first part
plt.scatter(x2[:5], x2[:5]**2,color='b', label='x2',alpha=0.5,s=10)
# for line two second part
plt.scatter(x2[4:],x2[4:]**2,color='g')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('This is title of graph' ,fontdict= {'fontname':'Arial' , 'fontsize' : 20 } )
plt.xticks([0,1,2,3])
plt.yticks([0,2,4,6,8,10])

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.stackplot(x, y1, y2, labels=['sin(x)', 'cos(x)'], colors=[ 'orange','blue'])
plt.xlabel('X')
plt.ylabel('Value')
plt.title('Stacked Area Plot')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
sizes = np.array([30, 20, 25, 25])
labels = ['A', 'B', 'C', 'D']
plt.pie(sizes, labels=labels, colors=['red', 'blue', 'green', 'yellow'], autopct='%.1f%%', startangle=30)
plt.title('Pie Chart')
plt.show()
```

```python

sizes = np.array([30, 20, 25, 25])
labels = ['A', 'B', 'C', 'D']
plt.pie(sizes, labels=labels, colors=['red', 'blue', 'green', 'yellow'], autopct='%.1f%%', startangle=30,explode=[0.1,0,0,0],shadow=True)
plt.title('Pie Chart')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.figure(figsize=(6,3),dpi=70)
plt.fill_between(x, y, color='lightblue', alpha=0.5, label='sin(x)')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Area Plot')
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.figure(figsize=(6,3),dpi=70)
plt.fill_between(x, y, color='lightblue', alpha=0.5, label='sin(x)')
plt.plot(x,y,'r.--')
plt.fill_between(x,np.cos(x), color='g', alpha=0.5, label='cos(x)')
plt.plot(x,np.cos(x),'b.--')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Area Plot')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import random as r
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = np.random.normal(0, 1, 1000)
plt.hist(data, bins=30, density=True, color='skyblue',alpha=0.7, label='Data')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram')
plt.legend()
plt.grid(True)
plt.show()
```

```python
data = np.random.normal(0, 1, 1000)
plt.hist(data, bins=30, density=True, color='skyblue',edgecolor='r',alpha=0.5, label='Data')
plt.hist(data, bins=30, density=True, color='blue',edgecolor='g',histtype='step',alpha=1, label='Data-|')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a= np.arange(0,15,1)
b= np.arange(0,15,1)

plt.step(a,b)
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.figure(figsize=(6,3),dpi=50)
plt.step(x,y,color='g',label='x,sin(x)')
# plt.step(x,y,color='r',where='pre')
plt.legend()
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.figure(figsize=(6,3),dpi=50)
# plt.step(x,y,color='g',label='x,sin(x)')
plt.step(x,y,color='r',where='post',label='x,sin(x)')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(6)
y = np.array([1, 2, 3, 2, 4, 3])
plt.figure(figsize=(8,3),dpi=100)
plt.subplot(1,)
plt.step(x,y,where='pre', color='g', label='Step Data')
plt.step(x,y,where='mid', color='r', label='Step Data')
plt.step(x,y,where='post', color='purple', label='Step Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Step Plot')
plt.legend()
plt.grid(True)
plt.show()
```

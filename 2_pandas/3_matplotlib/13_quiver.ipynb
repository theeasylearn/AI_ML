```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python

x = np.linspace(-2, 2, 10)
y = np.linspace(-2, 2, 10)
X, Y = np.meshgrid(x, y)
U = Y
V = -X
plt.figure(figsize=(6,3),dpi=80)
plt.quiver(X, Y, U, V, color='blue')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Quiver Plot')
plt.grid(True)
plt.show()
```
